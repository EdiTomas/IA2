```python
# Import Perceptron
from sklearn.neural_network import MLPClassifier

# Training set
#
x=[[0,2],[0,1],[2,0],[1,1],[2,2],[1,2],[2,1],[1,0]]
y = [1,1,0,0,0,0,0,0]




# Create model
model = MLPClassifier(activation='logistic', max_iter=100,
                      hidden_layer_sizes=(2,), alpha=0.0001,
                      solver='lbfgs')

# Train the model
model.fit(x,y)

# Predict Output
print('predictions:', model.predict(x))



```

```python
# Import Perceptron
from sklearn import preprocessing
from sklearn.neural_network import MLPClassifier

# Data creation


x1=[0,0,2,1,2,1,2,1]
x2=[2,1,0,1,2,2,1,0]
salida=[1,1,0,0,0,0,0,0]


# Creating labelEncoder
le = preprocessing.LabelEncoder()

# Converting string labels into numbers.
x1_encoded=le.fit_transform(x1)
x2_encoded=le.fit_transform(x2)
label=le.fit_transform(salida)


print(label)

# Combinig attributes into single listof tuples
features=list(zip(x1_encoded,x2_encoded))

# Create and fit the model
model = MLPClassifier(activation='logistic', max_iter=200,
                      hidden_layer_sizes=(10,), alpha=0.001,
                      solver='lbfgs')
model.fit(features, label)

#predicted = model.predict([[0.5,0.5]])
#predicted = model.predict(features)
#print("Predicted value:", predicted)


#predicted = model.predict([[2, 2]])
predicted = model.predict(features)
print("Predicted value:", predicted)
```
