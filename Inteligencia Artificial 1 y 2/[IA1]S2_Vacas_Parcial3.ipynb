```python

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
import numpy as np
# Obtener las características (atributos) y el objetivo (variable a predecir)

def regresionpolinomial(x_new,y_new,degree):
    x = np.array(x_new).reshape(-1, 1) # INDEPENDIENTE
    y = np.array(y_new) # DEPENDIENTE
    #Transformar las características a un polinomio de grado 2
    poly_transformer = PolynomialFeatures(degree, include_bias=False)
    X_poly = poly_transformer.fit_transform(x)

    # Crear y entrenar el modelo de regresión lineal polinomial
    model = LinearRegression()
    model.fit(X_poly, y)

    # Realizar las predicciones utilizando el modelo entrenado
    predicciones = model.predict(X_poly)

    # Calcular el coeficiente de determinación (R cuadrado)
    r2 = r2_score(y, predicciones)
    return r2
r2= regresionpolinomial(df['NO'],df['B'],2)
# Imprimir el valor de R cuadrado con 2 decimales sin redondear
print(f"El valor de R cuadrado del atributo B utilizando regresión polinomial de grado 2 es: {r2:.2f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Cargar el archivo pred.csv en un DataFrame
#data = pd.read_csv('pred.csv')
data = pd.read_csv('https://raw.githubusercontent.com/EdiTomas/Excel/master/pred.csv')

def regresionpolinomial_predicion(x_new,y_new,degree,correlativo):
    # Obtener las características (atributos) y el objetivo (variable a predecir)
    x = np.array(x_new).reshape(-1, 1) # INDEPENDIENTE
    y = np.array(y_new) # DEPENDIENTE

# Transformar las características a un polinomio de grado 2
    poly_transformer = PolynomialFeatures(degree, include_bias=False)
    X_poly = poly_transformer.fit_transform(x)

# Crear y entrenar el modelo de regresión lineal polinomial
    model = LinearRegression()
    model.fit(X_poly, y)

# Realizar la predicción para el correlativo 20
    correlativo_20 = np.array([[correlativo]])  # Convertir a un array bidimensional
    correlativo_20_poly = poly_transformer.transform(correlativo_20)
    prediccion = model.predict(correlativo_20_poly)
    return prediccion
# Imprimir la predicción con 2 decimales sin redondear

prediccion = regresionpolinomial_predicion(df['NO'],df['B'],2,30)
print(f"La predicción del atributo B para un correlativo 30 utilizando regresión polinomial de grado 2 es: {prediccion[0]:.3f}")

prediccion = regresionpolinomial_predicion(df['NO'],df['B'],2,50)
print(f"La predicción del atributo B para un correlativo 50 utilizando regresión polinomial de grado 2 es: {prediccion[0]:.3f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

# Cargar el archivo pred.csv en un DataFrame
#data = pd.read_csv('pred.csv')
data = pd.read_csv('https://raw.githubusercontent.com/EdiTomas/Excel/master/pred.csv')
df = pd.DataFrame(data)
# Obtener las características (atributos) y el objetivo (variable a predecir)
def regresionpolinomial_RMSE(x_new,y_new,degree):
       x = np.array(x_new).reshape(-1, 1) # INDEPENDIENTE
       y = np.array(y_new) # DEPENDIENTE
# Transformar las características a un polinomio de grado 2
       poly_transformer = PolynomialFeatures(degree, include_bias=False)
       X_poly = poly_transformer.fit_transform(x)

# Crear y entrenar el modelo de regresión lineal polinomial
       model = LinearRegression()
       model.fit(X_poly, y)

# Realizar las predicciones utilizando el modelo entrenado
       predicciones = model.predict(X_poly)

# Calcular el error cuadrático medio (RMSE)
       rmse = np.sqrt(mean_squared_error(y, predicciones))
       return rmse
# Imprimir el valor del RMSE con 2 decimales sin redondear


rmse = regresionpolinomial_RMSE(df['NO'],df['B'],2)
print(f"El valor del RMSE del atributo B utilizando regresión polinomial de grado 2 es: {rmse:.3f}")
```

```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB

# Cargar el archivo pred.csv en un DataFrame
#data = pd.read_csv('pred.csv')
#x = data[['A', 'B', 'C', 'D']]  # Características: A, B, C, D

def Gausiano_Prediccion(valores):
    data = pd.read_csv('https://raw.githubusercontent.com/EdiTomas/Excel/master/pred.csv')
    # Obtener las variables de entrada (A, B, C, D) y la variable de salida (E)
    A =data['A']
    B =data['B']
    C =data['C']
    D =data['D']
    x=list(zip(A,B,C,D))
    y = data['E']  # Objetivo: E
    # Crear y entrenar el clasificador GaussianNB
    classifier = GaussianNB()
    classifier.fit(x, y)
    # Valores de entrada para la predicción
    valores_entrada = [valores] #[[5, 500, 200, False]]
    # Realizar la predicción
    prediccion = classifier.predict(valores_entrada)
    return prediccion
# Imprimir la predicción


prediccion = Gausiano_Prediccion([10,20,400,True])
print(f'La predicción para los valores de entrada [10,20,400,True] es: {prediccion[0]}')
prediccion = Gausiano_Prediccion([5, 500, 200, False])
print(f'La predicción para los valores de entrada [5, 500, 200, False] es: {prediccion[0]}')
prediccion = Gausiano_Prediccion([100,100,550,True])
print(f'La predicción para los valores de entrada [100,100,550,True] es: {prediccion[0]}')
prediccion = Gausiano_Prediccion([10, 20, 400, True])
print(f'La predicción para los valores de entrada [10, 20, 400, True] es: {prediccion[0]}')

prediccion = Gausiano_Prediccion([100,0,100,False])
print(f'La predicción para los valores de entrada [100,0,100,False] es: {prediccion[0]}')
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

# Cargar el archivo pred.csv en un DataFrame
#data = pd.read_csv('pred.csv')
data = pd.read_csv('https://raw.githubusercontent.com/EdiTomas/Excel/master/pred.csv')

# Obtener las características (atributos) y el objetivo (variable a predecir)
X = data[['NO']]  # Característica: Correlativo
y = data['B']  # Objetivo: B

grados = range(1, 10)  # Probar grados del polinomio desde 1 hasta 9
mejor_grado = 0
mejor_r2 = -1

for grado in grados:
    # Transformar las características a un polinomio del grado actual
    poly_transformer = PolynomialFeatures(degree=grado, include_bias=False)
    X_poly = poly_transformer.fit_transform(X)

    # Crear y entrenar el modelo de regresión lineal
    model = LinearRegression()
    model.fit(X_poly, y)

    # Realizar las predicciones utilizando el modelo entrenado
    predicciones = model.predict(X_poly)

    # Calcular el coeficiente de determinación (R cuadrado)
    r2 = r2_score(y, predicciones)

    # Actualizar el grado y R cuadrado si se encuentra un valor mejor
    if r2 > mejor_r2:
        mejor_r2 = r2
        mejor_grado = grado

print(f"El grado más adecuado según el R cuadrado para el atributo B es: {mejor_grado}")
```
