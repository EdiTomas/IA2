```python
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Datos de la tabla
x = np.array([100, 90, 70, 40])
y = np.array([45, 52, 58, 62])

# Inicializar variables para el mejor grado
mejor_grado = 1
mejor_rmse = float('inf')  # Inicializar con infinito
mejor_r2 = -float('inf')   # Inicializar con menos infinito

# Probar diferentes grados polinomiales
for grado in range(1, 11):  # Prueba grados del 1 al 10
    polynomial_features = PolynomialFeatures(degree=grado)
    X_TRANSF = polynomial_features.fit_transform(x.reshape(-1, 1))

    model = linear_model.LinearRegression()
    model.fit(X_TRANSF, y)
    Y_NEW = model.predict(X_TRANSF)

    # Calcular métricas
    rmse = np.sqrt(mean_squared_error(y, Y_NEW))
    r2 = r2_score(y, Y_NEW)

    # Actualizar si se encuentra un mejor grado
    if rmse < mejor_rmse and r2 > mejor_r2:
        mejor_grado = grado
        mejor_rmse = rmse
        mejor_r2 = r2

# Imprimir el grado más adecuado
print("El grado más adecuado es:", mejor_grado)
print("RMSE mínimo para ese grado:", mejor_rmse)
print("R2_score máximo para ese grado:", mejor_r2)
```

```python
from sklearn.metrics import silhouette_score
from matplotlib import pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
X = np.array([[217,	139,	80,	0.33,	4,	35,	3.7], #a
              [197,	197,	60,	0.15,	3.5,	30,	3.7], #b
              [240,	176,	60,	0.15,	3,	20,	3.7],#c
              [240,	186,	40,	0.16,	2,	15,	3.7]])#d


sse = {}
for k in range(1, 5):
    kmeans = KMeans(n_clusters=k, max_iter=2000).fit(X)
    sse[k] = kmeans.inertia_
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()

for n_cluster in range(2, 4):
    kmeans = KMeans(n_clusters=n_cluster).fit(X)
    label = kmeans.labels_
    sil_coeff = silhouette_score(X, label)
    print("For c={}, The Silhouette Coef is {}".format(n_cluster, sil_coeff))
```
