```python
import numpy as np

# Datos de los casos
datos = {
    'A': {'ganancia': 5, 'costo': 3},
    'B': {'ganancia': 6, 'costo': 4},
    'C': {'ganancia': 2, 'costo': 1},
    'E': {'ganancia': 9, 'costo': 3},
    'F': {'ganancia': 8, 'costo': 5},
    'G': {'ganancia': 9, 'costo': 3}
}

# Población inicial
poblacion = ['ABC', 'CEF', 'EFG']

# Función para evaluar la ganancia total y el costo total de una mochila
def evaluar_mochila(individuo):
    ganancia_total = sum(datos[punto]['ganancia'] for punto in individuo)
    costo_total = sum(datos[punto]['costo'] for punto in individuo)
    return ganancia_total, costo_total

# Función para seleccionar los mejores padres
def seleccionar_padres(poblacion, num_padres):
    puntuaciones = [evaluar_mochila(individuo)[0] for individuo in poblacion]
    indices_padres = np.argsort(puntuaciones)[::-1][:num_padres]
    return [poblacion[i] for i in indices_padres]

# Función para cruzar padres
def cruzar_padres(padre1, padre2):
    punto_cruce = np.random.randint(1, len(padre1))
    hijo1 = padre1[:punto_cruce] + padre2[punto_cruce:]
    hijo2 = padre2[:punto_cruce] + padre1[punto_cruce:]
    return hijo1, hijo2

# Función para reemplazar la población
def reemplazar_poblacion(poblacion, hijos):
    poblacion.extend(hijos)
    puntuaciones = [evaluar_mochila(individuo)[0] for individuo in poblacion]
    indices_mejores = np.argsort(puntuaciones)[::-1][:len(poblacion)]
    return [poblacion[i] for i in indices_mejores]

# Evolución hasta la cuarta generación
ganancias_generacion = []

for _ in range(4):
    ganancias_generacion.append(sum(evaluar_mochila(individuo)[0] for individuo in poblacion))
    padres = seleccionar_padres(poblacion, 2)
    hijos = cruzar_padres(padres[0], padres[1])
    poblacion = reemplazar_poblacion(poblacion, hijos)

# Calcular la convergencia en porcentaje
convergencia_porcentaje = ((ganancias_generacion[2] - ganancias_generacion[1]) / ganancias_generacion[1]) * 100

# Imprimir el resultado
print(f"La convergencia en la tercera generación es aproximadamente del {convergencia_porcentaje:.2f}%.")

# Convertir a porcentaje entero sin decimales
convergencia_porcentaje_entero = int(convergencia_porcentaje)

# Imprimir el resultado sin decimales
print(f"La convergencia en la tercera generación es del {convergencia_porcentaje_entero}%.")
```
