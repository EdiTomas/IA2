```python
from sklearn.cluster import KMeans

import numpy as np
from sklearn.manifold import MDS
from matplotlib import pyplot as plt

X = np.array([[217,	139,	80,	0.33,	4,	35,	3.7], 
              [197,	197,	60,	0.15,	3.5,	30,	3.7],
              [240,	176,	60,	0.15,	3,	20,	3.7], 
              [240,	186,	40,	0.16,	2,	15,	3.7]])

#mds = MDS(n_clusters=3,random_state=0)
#X_transform = mds.fit_transform(X)
#print(X_transform)

#colors = ['k', 'r', 'g', 'b']
#size = [64, 64, 64, 64]
#plt.axis('equal')
#plt.scatter(X_transform[:,0], X_transform[:,1], s=size, c=colors)
#plt.title('2D')
#plt.show()




kmeans = KMeans(n_clusters=3,  random_state=0)
kmeans.fit(X)
print(kmeans.cluster_centers_)
print(kmeans.labels_)


kmeans2D = KMeans(n_clusters=3, random_state=0)
kmeans2D.fit(X)
print(kmeans2D.cluster_centers_)
print(kmeans2D.labels_)
plt.axis('equal')
plt.scatter(X[:,0],
            X[:,1], 
            c=kmeans2D.labels_, 
            cmap='rainbow',)


```
