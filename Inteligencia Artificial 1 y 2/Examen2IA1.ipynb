```python
import heapq

# Función para calcular la distancia de Manhattan entre dos puntos (a, b) y (x, y)
def manhattan_distance(a, b, x, y):
    return abs(a - x) + abs(b - y)

# Función para calcular la distancia de Manhattan total de un estado
def total_manhattan_distance(state):
    distance = 0
    for i in range(2):
        for j in range(3):
            if state[i][j] != 0:
                distance += manhattan_distance(i, j, (state[i][j] - 1) // 3, (state[i][j] - 1) % 3)
    return distance

# Función para encontrar la solución utilizando el algoritmo Best First
def best_first(initial_state):
    # Cola de prioridad (usamos una lista de tuplas para ordenar por distancia)
    priority_queue = [(0, initial_state)]

    while priority_queue:
        distance, current_state = heapq.heappop(priority_queue)

        if current_state == [[1, 2, 3], [4, 5, 0]]:
            return current_state  # Hemos llegado al estado final

        # Encontrar la posición del espacio en blanco
        for i in range(2):
            for j in range(3):
                if current_state[i][j] == 0:
                    x, y = i, j

        # Generar posibles movimientos
        for dx, dy in [(0, 1), (1, 0), (0, -1), (-1, 0)]:
            new_x, new_y = x + dx, y + dy
            if 0 <= new_x < 2 and 0 <= new_y < 3:
                new_state = [list(row) for row in current_state]
                new_state[x][y], new_state[new_x][new_y] = new_state[new_x][new_y], new_state[x][y]
                heapq.heappush(priority_queue, (total_manhattan_distance(new_state), new_state))

# Estado inicial
initial_state = [[2, 4, 3], [1, 0, 5]]

# Resolver el puzzle
solution = best_first(initial_state)

# Imprimir la solución
for row in solution:
    print(row)
```
