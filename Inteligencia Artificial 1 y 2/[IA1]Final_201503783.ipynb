```python
from collections import deque

# Definir la Tabla No. 2 como una lista de listas
tabla = [
    [100, '-', 45, 'puff', '-', '-', '-'],
    [90, 20, 'min', '-', 52, 'puff', '-'],
    [70, 40, 'min', 20, 'min', '-', 58, 'puff', '-'],
    [40, 60, 'min', 40, 'min', 20, 'min', '-', 62, 'puff'],
    [0, 80, 'min', 60, 'min', 40, 'min', 20, 'min', '-']
]

def generar_nodos(tabla, nivel_max):
    nodos_generados = 0

    # Utilizar una cola (deque) para realizar la búsqueda por anchura
    cola = deque([(0, 0)])  # Iniciar desde el nodo raíz (0, 0)

    while cola:
        estado, nivel = cola.popleft()  # Obtener el siguiente nodo de la cola
        nodos_generados += 1

        if nivel < nivel_max:
            # Obtener los hijos del nodo actual en el siguiente nivel
            fila = tabla[nivel]
            for i in range(estado, len(fila)):
                if fila[i] != '-':
                    cola.append((i + 1, nivel + 1))  # Agregar el nodo hijo a la cola

    return nodos_generados

# Ejecutar el algoritmo de búsqueda por anchura con limitación en el nivel 3
num_nodos_generados = generar_nodos(tabla, nivel_max=3)
print("Número de nodos generados:", num_nodos_generados)
```

```python
import queue

# Definir los datos de la Tabla No. 2
tabla = [
    [100, "-", 45, "-", "-", "-"],
    [90, 20, "-", 52, "-", "-"],
    [70, 40, 20, "-", 58, "-"],
    [40, 60, 40, 20, "-", 62],
    [0, 80, 60, 40, 20, "-"]
]

# Función de costo acumulado en dólares
def costo_acumulado(costo_padre, costo_hijo):
    if costo_hijo == "-":
        return costo_padre
    else:
        return costo_padre + costo_hijo

# Definir la meta de generar 135 o más puff
meta = 135

# Cola de prioridad para almacenar los nodos del árbol
cola = queue.PriorityQueue()

# Nodo inicial
nodo_inicial = (0, 0, [], 0.0)
cola.put(nodo_inicial)

# Costo acumulado del nodo que generó 155 puff
costo_nodo_155 = 0.0

# Ejecutar el algoritmo de búsqueda por costo uniforme
while not cola.empty():
    costo_padre, nivel, ruta, costo_dolares = cola.get()

    # Verificar si se ha alcanzado la meta
    if nivel >= meta:
        costo_nodo_155 = costo_dolares
        break

    # Generar nodos hijos
    for i in range(0):
        costo_hijo = tabla[nivel][i]
        if costo_hijo != "-":
            nueva_ruta = ruta + [(costo_hijo,)]
            nuevo_costo = costo_acumulado(costo_padre, costo_hijo)
            nuevo_costo_dolares = costo_dolares + float(costo_hijo)
            nuevo_nodo = (nuevo_costo, nivel + 1, nueva_ruta, nuevo_costo_dolares)
            cola.put(nuevo_nodo)

# Imprimir el costo en dólares del nodo que generó 155 puff
formatted_costo_nodo_155 = "{:.2f}".format(costo_nodo_155)
print("Costo en dólares del nodo que generó 155 puff:", formatted_costo_nodo_155)
```

```python
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

x = np.array([100, 90, 70, 40]).reshape(-1, 1)
y = np.array([45, 52, 58, 62])
nb_degree = 5
polynomial_features = PolynomialFeatures(degree = nb_degree)
X_TRANSF = polynomial_features.fit_transform(x)

model = linear_model.LinearRegression()
model.fit(X_TRANSF, y)
Y_NEW = model.predict(X_TRANSF)
rmse = np.sqrt(mean_squared_error(y,Y_NEW))
r2 = r2_score(y,Y_NEW)
print('RMSE: ', rmse)
print('R2: ', r2)

x_new_min = 0
x_new_max = 100

X_NEW = np.linspace(x_new_min, x_new_max,100)
X_NEW = X_NEW[:,np.newaxis]
X_NEW_TRANSF = polynomial_features.fit_transform(X_NEW)

Y_NEW = model.predict(X_NEW_TRANSF)
plt.scatter(x, y,  color='black')
plt.plot(X_NEW, Y_NEW, color='coral', linewidth=3)
plt.grid()
plt.xlim(x_new_min,x_new_max)
title = 'Degree = {}; RMSE = {}; R2 = {}'.format(nb_degree, round(rmse,2),
                                                 round(r2,2))
plt.title("Polynomial Linear Regression using scikit-learn and python 3 \n " +
          title, fontsize=10)
plt.xlabel('x')
plt.ylabel('y')
```

```python

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

# Datos de la Tabla No. 2
#x = np.array([100, 90, 70, 40])  # Valores en el eje x
#y = np.array([100, 45, 52, 58])  # Valores en el eje y (puff)

x = np.array([100, 90, 70, 40]).reshape(-1, 1)
y = np.array([45, 52, 58, 62])

# Inicializar los valores de RMSE mínimo y R2 Score máximo
rmse_min = float('inf')
r2_score_max = float('-inf')
best_degree = None

# Probar diferentes grados de regresión polinomial
for degree in range(1, 11):
    # Crear características polinomiales del grado actual
    poly_features = PolynomialFeatures(degree=degree)
    x_poly = poly_features.fit_transform(x.reshape(-1, 1))

    # Entrenar el modelo de regresión lineal
    model = LinearRegression()
    model.fit(x_poly, y)

    # Realizar la predicción
    y_pred = model.predict(x_poly)

    # Calcular el RMSE y R2 Score
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    r2 = r2_score(y, y_pred)

    # Actualizar los valores mínimos y máximos si corresponde
    if rmse < rmse_min:
        rmse_min = rmse
        best_degree_rmse = degree
    if r2 > r2_score_max:
        r2_score_max = r2
        best_degree_r2 = degree

print(f"El grado de regresión polinomial más adecuado considerando RMSE mínimo es: {best_degree_rmse}")
print(f"El grado de regresión polinomial más adecuado considerando R2 Score máximo es: {best_degree_r2}")

```

```python
import heapq

# Datos de la Tabla No. 2
tabla = [
    [100, "-", "-", "-", "-"],
    [90, 20, "-", 52, "-"],
    [70, 40, 20, "-", 58],
    [40, 60, 40, 20, "-"],
    [0, 80, 60, 40, 20]
]

# Variables para controlar el algoritmo
niveles = 3  # Limitación de nivel en el árbol
objetivo_puff = 135  # Puff objetivo
max_puff = 0  # Variable para almacenar el máximo número de puff encontrado

# Crear la cola de prioridad (min-heap) para el algoritmo de costo uniforme
cola = [(0, 0, 0, [])]  # (costo acumulado, nivel, puff actual, camino)

while cola:
    costo_acumulado, nivel, puff, camino = heapq.heappop(cola)

    # Verificar si se alcanzó el nivel objetivo y se generaron suficientes puff
    if nivel == niveles and puff >= objetivo_puff:
        print("Solución encontrada:")
        print("Costo acumulado en dólares:", costo_acumulado)
        print("Puff generados:", puff)
        print("Camino:", camino)
        break

    # Verificar si se alcanzó el máximo número de puff encontrado
    if puff > max_puff:
        max_puff = puff

    # Generar los posibles estados en el siguiente nivel
    for i in range(len(tabla[nivel])):
        if tabla[nivel][i] != "-":
            costo = int(tabla[nivel][i])
            nuevo_costo_acumulado = costo_acumulado + costo
            nuevo_puff = puff + costo

            nuevo_camino = camino + [(costo, puff)]

            heapq.heappush(cola, (nuevo_costo_acumulado, nivel + 1, nuevo_puff, nuevo_camino))

print("Máximo número de puff encontrado en un estado:", max_puff)
```
